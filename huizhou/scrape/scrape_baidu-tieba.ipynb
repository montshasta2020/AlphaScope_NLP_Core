```python
def urllib_get_page_with_retry(url):
    stuff=None
    max_attempts=5
    r=0
    while stuff is None and r<max_attempts:
        try:
            response = urllib2.urlopen(url, timeout=500)
            stuff = response.read()
        except urllib2.URLError:
            r=r+1
            print "Re-trying, attempt -- ",r
            time.sleep(5)
            pass
    return stuff
```
