```python
res = []
cnt = 0
pos_cnt = 0
neg_cnt = 0
```

```python
def get_columns3():
    return ('url', 'token title', 'flag') 
def get_empty_columns3():
    return ('', '', '') 
```

```python
idx = 0
new_df = pd.DataFrame(columns = get_columns3())
```

```python
for i in range(0, len(df_no_dup)):
    t0 = df_no_dup.loc[i]['token title']
    t1 = df_no_dup.loc[i]['flag']
    url = df_no_dup.loc[i]['url']
    #print i, url
    if url in ref_df['url'].values:
        print 'skip', cnt
        cnt += 1
    else:
        new_df.loc[idx] = get_empty_columns3()
        new_df.loc[idx]['url'] = url
        new_df.loc[idx]['token title'] = t0 
        if t1 == 'pos':
            pos_cnt += 1
            new_df.loc[idx]['flag'] = 1
        elif t1 == 'neg':
            neg_cnt += 1
            new_df.loc[idx]['flag'] = 0
        else:
            print 'Missing ...'
        temp = (t0, t1)
        #print temp
        res.append(temp)
        idx += 1
```
