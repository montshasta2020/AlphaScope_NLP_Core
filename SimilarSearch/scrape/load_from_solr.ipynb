```python
idx = 0
#filter the titles from Chinese Medicine
def map_strip_titles6(x):
    global idx
    #print idx
    #print x
    x = x.replace('- 中医中药秘方网', '')
    x = x.replace('>>中医中药秘方网', '')
    x = x.replace('- 民间中医网 -', '')
    
    
    t = x.split('_')
    s = t[0]
    r = s.split('|')
    s = r[0]
    idx += 1
    #print s
    return s
```

```python
idx = 0
#filter the titles from Chinese Medicine
def map_strip_titles3(x):
    global idx
    #print idx
    #print x
    x = x.replace('- Wikipedia', '')
    x = x.replace('Wikipedia:', '')
    x = x.replace('- CNN.com', '')
    x = x.replace('- Wikibon', '')
    x = x.replace('- AnalyticBridge', '')
    x = x.replace('- Data Science Central', '')
    x = x.replace('« Byte Mining', '')
    x = x.replace('- Enterprise Apps Tech News', '')
    x = x.replace('- comScore, Inc', '')
    #x = x.replace('- comScore, Inc', '')
    
    
    
    t = x.split('_')
    s = t[0]
    r = s.split('|')
    s = r[0]
    idx += 1
    #print s
    return s
```

```python
def map_strip_titles2(x):
    global idx
    #print idx
    #print x
    x = x.replace('-故园徽州网 - Powered by Discuz!', '')
    x = x.replace('婺源论坛 - 江西婺源旅游 - 婺源老家旅游论坛第一门户', '')
    x = x.replace('-中国华文教育网', '')
    x = x.replace('-中安在线-徽文化', '')
    x = x.replace('中国华文教育网－', '')

    
    t = x.split('_')
    s = t[0]
    idx += 1
    #print s
    return s
```
