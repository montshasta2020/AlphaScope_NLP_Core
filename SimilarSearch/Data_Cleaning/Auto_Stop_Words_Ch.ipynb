```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
count_vectorizer = CountVectorizer(min_df=1)
term_freq_matrix = count_vectorizer.fit_transform(tot_words)
dictionary = count_vectorizer.get_feature_names()
```

```python
print len(dictionary)
```

```python
print term_freq_matrix.shape[0], term_freq_matrix.shape[1]
```

```python
freq_mat = np.zeros((term_freq_matrix.shape[0],term_freq_matrix.shape[1]))
var_mat = np.zeros((term_freq_matrix.shape[0],term_freq_matrix.shape[1]))
```

```python
term_freq_matrix_dense = term_freq_matrix.todense()
```

```python
print term_freq_matrix_dense[0, 1], freq_mat[0, 0]
```

```python
for j in range(0, term_freq_matrix.shape[0]):
    for k in range(0, term_freq_matrix.shape[1]):
        #print j, k
        freq_mat[j, k] = 1.0*term_freq_matrix_dense[j, k]/df.iloc[j]['words_len']
```

```python
freq_mean = np.zeros(term_freq_matrix.shape[1])    
freq_var = np.zeros(term_freq_matrix.shape[1])    
```

```python
freq_mat.sum(0)
```

```python
freq_mean = freq_mat.sum(axis=0)*1.0/term_freq_matrix.shape[0]
```

```python
for j in range(0, term_freq_matrix.shape[0]):
    for k in range(0, term_freq_matrix.shape[1]):
        #print j, k
        var_mat[j, k] = (freq_mat[j, k] - freq_mean[k])*(freq_mat[j, k] - freq_mean[k])
```

```python
freq_var = var_mat.sum(axis=0)*1.0/term_freq_matrix.shape[0]
```

```python
sat = freq_mean/freq_var
```

```python
ret = np.argsort(sat)
```
