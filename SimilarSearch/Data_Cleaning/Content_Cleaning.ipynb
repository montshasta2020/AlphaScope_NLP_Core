```python
import pandas as pd
from urlparse import urlparse
import tldextract
from collections import Counter
from time import time
```

```python
#df = pd.read_csv('/Users/nhu2000/projects/SimilarSearch/data/clean/huizhou-web-slim_v0_1.csv')
df = pd.read_csv('/Users/nhu2000/projects/SimilarSearch/data/segments/huizhou_Solr_v1.csv')
```

```python
#find the domain and host
import tldextract
df['domain']  = df.url.map(lambda x:tldextract.extract(x).domain + str('.') + tldextract.extract(x).suffix)
```

```python
df.info()
```

```python
def get_clean_content(t, cut = 30):
    s = t.decode('utf-8')
    s = s.split(' ')
    
    ret = [a for a in s if len(a) > cut]
    ret0 = '\n'.join(ret)
    #print ret0
    if len(ret) > 0:
        return '\n'.join(ret)
    else:
        return None
```

```python
df['content'] = df.content.map(lambda x: get_clean_content(x))
```

```python
df = df.dropna(subset =['content'])
```
