```python
def parse_json(data):    
    refs = []
    missing_ref = []
    
    tot = int(data['result']['hits']['@total'])
    
    print ('tot----', tot)
    for k in range(tot):
        raw_ref = data['result']['hits']['hit'][k]
        
        #print('reference cnt', k, raw_ref)
        
        if 'authors' not in raw_ref['info']:
            print('missing reference', k, raw_ref)
            missing_ref.append(raw_ref['info']['title'])
            continue
        
        authors = raw_ref['info']['authors']['author']
        title = raw_ref['info']['title']
        year = raw_ref['info']['year']

        one_ref = {'names': [], 'year': [], 'title': []}
        
        one_ref['year'] = year
        one_ref['title'] =  title
        
        names = []
        
        if type(authors) == dict:
            one_ref['names'] = authors['text']
        else:        
            for l in range(len(authors)):
                #print (len(authors), l)
                names.append(authors[l]['text'])

            one_ref['names'] = ', '.join(map(str, names))
            
        refs.append(one_ref)
    
    return refs, missing_ref
```

```python
root_dir = './dblp_combined/*.json'
all_refs = []
all_miss = []
for fn in glob.glob(root_dir):
    
    f = open(fn)
    data = json.load(f)
    
    ret, miss = parse_json(data)
    
    all_refs = all_refs + ret
    all_miss = all_miss + miss
```
