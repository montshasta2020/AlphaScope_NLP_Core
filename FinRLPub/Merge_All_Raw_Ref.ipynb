```python
import glob
import pandas as pd

cnt = 0
for fn in glob.glob('References_2/*.csv'):
    df = pd.read_csv(fn)
    if cnt == 0:
        merged_df = df[['raw_ref']]
    else:
        merged_df = pd.concat([merged_df, df[['raw_ref']]])
    
    cnt += 1
    
    print(cnt, merged_df.shape, df.columns)
```
