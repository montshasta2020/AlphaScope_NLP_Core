```python
import pandas as pd
import re
```

```python
df = pd.read_csv('All_References_03.csv')
```

```python
df.head()
```

```python
clearned_raw_ref = []

ss = df['raw_ref'].values

cnt = 0
cnt1 = 0
for k in range (len(ss)):
    #match = re.match(r'.*([1-3][0-9]{3})', ss[k][0])
    match = re.search(r'[12]\d{3}', ss[k])

    #print(match)
    if match != None:
        
        year = int(match.group(0))
        if 1900 <= year <= 2022:
            print ('Found', cnt1, year)
            
            clearned_raw_ref.append(ss[k][2:-2])

            cnt1 += 1
            #print(ss[k])
    else:
        #print(cnt, ss[k])
        cnt += 1
```

```python
out_df = pd.DataFrame(clearned_raw_ref, columns=['raw_ref'])
```

```python
out_df.to_csv('Cleaned_Ref_03.csv')
```
