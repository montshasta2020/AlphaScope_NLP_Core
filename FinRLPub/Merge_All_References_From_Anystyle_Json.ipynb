```python
cnt = 0
cols = ['date', 'author', 'title']
init_df = False
for fn in glob.glob('./References/*.json'):
    if os.path.getsize(fn) > 0:
        data_df= pd.read_json(fn, orient=str)
        
        #print ('data_df', data_df.shape)
#        print(data_df.shape, data_df.columns)

        if cols[0] not in data_df.columns:
#            print('Missing date')
            continue

        if cols[1] not in data_df.columns:
#            print('Missing author')
            continue
    
        if cols[2] not in data_df.columns:
#            print('Missing author')
            continue
            
        sub_df = data_df[cols]
        
        #print (sub_df.shape, data_df.shape)
        
        if init_df == False:
            print('Init')
            merged_df = sub_df
            init_df = True
        else:
            merged_df = pd.concat([merged_df, sub_df])
            print(cnt, sub_df.shape, merged_df.shape)
    else:
        print('-------find empty', cnt, fn)
        
    cnt += 1
```

```python
def merge_names(data):
    names = ''
    
    if type(data) == dict:
        d = data
        if 'family' in d.keys() and 'given' in d.keys():
                names += d['family'] + ' ' + d['given']                 
    else:    
        for k in range(len(data)):
            d = data[k]
            if 'family' in d.keys() and 'given' in d.keys():
                if k == len(data) - 1:
                    names += d['family'] + ' ' + d['given'] 
                else:
                    names += d['family'] + ' ' + d['given'] + ', '
    return names
```

```python
cnt = 0

from dateutil.parser import parse
import re

for k in range(dropNA_df.shape[0]):
    one_row = dropNA_df.iloc[k]
    
    
    if len(one_row['date']) == 1 and len(one_row['title']) == 1:
        if len(one_row['date'][0]) > 4:
#            ys = int(parse(one_row['date'][0], fuzzy=True).year)
            match = re.search('\d{4}', one_row['date'][0])
            if match == None:
                ys = 2022
            else:
                ys = int(match.group(0))
            print('match year', one_row['date'][0], ys)
        else:
            s = one_row['date'][0]
            
            if s == 'May':
                ys = 2009
            elif s[-1] == 'l':
                 ys = 1991
            elif s == 'XXXX':
                ys = 2021
            else:                
                match = re.search('\d{4}', one_row['date'][0])
                if match != None:
                    ys = int(s)
                else:
                    ys = None

        if ys != None and len(one_row['author']) > 0:
            new_df.at[cnt, 'year'] = ys
            new_df.at[cnt, 'title'] = str(one_row['title'][0])
            new_df.at[cnt, 'authors'] = merge_names(one_row['author'][0])
        
        print (k, ys, new_df.shape)
        cnt += 1
```
