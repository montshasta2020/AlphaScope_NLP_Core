```python
import pandas as pd
from selenium import webdriver 
from bs4 import BeautifulSoup
import time

from datetime import date, timedelta, datetime
import os.path

from dateutil.parser import parse
```

```python
fname0 = './data/home_urls.csv'
fname1 = './data/general/google_scape_all.csv'

df0 = pd.read_csv(fname0)

browser = webdriver.Firefox() 
```

```python
def get_columns():
    return ('dt', 'source', 'domain', 'url', 'title', 'summary', 'image', 'brief', 'body', 'raw_page', 'from') 
def get_empty_columns():
    return ('', '', '', '', '', '', '', '', '', '', '') 
def get_empty_columns1():
    return ('', '', '', '', '', '', '', '', '', '', '','') 
```

```python
def is_date(string):
    try: 
        parse(string)
        return True
    except ValueError:
        return False
```

```python
if os.path.isfile(fname1): 
    df = pd.read_csv(fname1)
    idx = len(df.index)
else:
    df = pd.DataFrame(columns = get_columns())
    idx = 0
```

```python
print len(df)
```

```python
def is_notwanted (txt):
    if ('.com' in txt or '.cn' in txt or '.net' in txt):
        return True
    else:
        return False
```

```python
for item in df0['home']:
    link = 'https://www.google.com/#q=' + item +' %E5%BE%BD%E6%96%87%E5%8C%96&filter=0&num=100'    
    #link = 'https://www.google.com/#q=' + item +' %E5%BE%BD%E6%96%87%E5%8C%96&num=100'
    print link
    browser.get(link)
    time.sleep(5)
    
    soup = BeautifulSoup(browser.page_source)
    sub_texts = soup.findAll('div', {'class':'srg'})

    for txt in sub_texts:

        #set title and url    
        links = txt.findAll('div', {'class': 'g'})
        
        for link in links:
            #check if it is already existed
            #tmp_df = df[df['url'] == link.a['href']
            t0 = abs(sum(df['url'].str.find(str(link.a['href'])))) 

            if(t0 == len(df)  and is_notwanted(link.a.text) == False ):
                #print link.a.text
                df.loc[idx] = get_empty_columns()

                #set title and url    
                df.loc[idx]['url'] = link.a['href']
                df.loc[idx]['title'] = link.a.text

                #set image source
                if txt.a.img:
                    df.loc[idx]['image'] = txt.a.img['src'] 

                #split between source and time with '-'
                t1 = link.div.text
                t2 = t1.split('-')

                #print t2[0]

                #if "ago" in t2[0]:
                #    df.loc[idx]['dt'] = t0                
                #elif is_date(t2[0]):
                #    print t2[0]
                #    df.loc[idx]['dt'] = pd.to_datetime(t2[0]).to_pydatetime().date()

                df.loc[idx]['source'] = item

                t3 = link.find('span', {'class': 'st'})
                df.loc[idx]['brief'] = t3.text

                #print txt, t2
                df.loc[idx]['from'] = 'google'

                #load_one_page(df, idx, link.a['href'])
                idx += 1 
            
    time.sleep(10)
```

```python
df.to_csv(fname1, encoding='utf-8', index=False)
```
